```python
# Data Modeling & Merge Strategy
# 1. Order-Level Analysis Table

order_items_enriched = (
                         items
                        .merge(orders, on='order_id', how='inner')
                        .merge(products, on='product_id', how='left')
)

order_items_enriched.head()
```

```python
order_items_enriched = (
    order_items_enriched
    .drop(columns=['user_id_y'])
    .rename(columns={'user_id_x': 'user_id'})
)
```

```python
# Now we are moving to find insights
# 1. Total Revenue
total_revenue = order_items_enriched['calculated_total'].sum()
total_revenue
```

```python
# 2. Total_orders
total_orders = order_items_enriched['order_id'].nunique()
total_orders
```

```python
# 3. Average Order Value (AOV)
AOV = total_revenue/total_orders
AOV
```

```python
# 4. Total Quantity Sold
total_quantity = order_items_enriched['quantity'].sum()
total_quantity
```

```python
# 5. Revenue by Category
revenue_by_category = (order_items_enriched.groupby('category')['calculated_total'].sum().sort_values(ascending=False))
revenue_by_category
```

```python
# 6. Top 10 Products by Revenue
Top_10 = (order_items_enriched
    .groupby('product_name')['calculated_total']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
Top_10
```

```python
# Now we analyze customer behavior(Customer Metrics)
# 1. Total customers(How many orders each customer made.)
customer_summary = (
    customer_orders
    .groupby('user_id')
    .agg(
        total_orders = ('order_id', 'nunique'),
        total_spent = ('total_amount', 'sum')
    )
)

customer_summary
```

```python
# 2. Customers with ≥ 2 orders (repeat customers)
repeat_customers = customer_summary[customer_summary['total_orders'] >= 2]
repeat_customers
```

```python
# 3. Repeat customer rate
repeat_customer_rate = len(repeat_customers)/ len(customer_summary)
repeat_customer_rate
```

```python
# 4. Top 10 customers by total spend
top_customers = customer_summary.sort_values(
    'total_spent', ascending=False
).head(10)

top_customers
```

```python
# 5. How much revenue comes from loyal (repeat) customers vs new/one-time customers?
# steps :
# i) Identify number of orders per user : we did this as customer_summary

# ii) Classify each user as: Repeat, One-time

# iii) Sum total revenue for each group

# iv) Compare contribution

# step-2
customer_summary['customer_type'] = np.where(
    customer_summary['total_orders'] >= 2, 
    'Repeat',
    'One-Time'
)
```

```python
# step-3 & 4
revenue_by_customer_type = (
    customer_summary
    .groupby('customer_type')['total_spent']
    .sum()
)
revenue_by_customer_type
```

```python
# 6. Revenue share %
revenue_share = (
    revenue_by_customer_type / revenue_by_customer_type.sum() * 100
).round(2)

revenue_share
```
